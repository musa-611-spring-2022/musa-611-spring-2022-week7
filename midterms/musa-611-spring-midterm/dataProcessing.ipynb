```python
import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
traps = pd.read_json('data/traps.json')
traps.iloc[2]
```

```python
traps['location'][55]
```

```python
trapLct = []
for i in np.arange(0, len(traps)):
  if traps['location'][i] not in trapLct:
    trapLct.append(traps['location'][i]) 
```

```python
trapLct = pd.DataFrame(trapLct[1:])[['latitude','longitude']]
```

```python
trapLct.to_json('data/trap_location.json', orient="records")
```
